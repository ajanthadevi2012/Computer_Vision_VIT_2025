```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('/content/burano.jpg')

plt.imshow(img)
```

```python
# Convert the image into RGB
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
def display(img, cmap = None):
    fig = plt.figure(figsize = (10,10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap = cmap)
```

# color mode

```python
img_gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
display(img_gray, cmap = 'gray')
```

```python
fig,axs = plt.subplots(nrows = 1, ncols = 3, figsize = (20,10))
for i in range (0, 3):
    ax = axs[i]
    ax.imshow(img_rgb[:,:,i], cmap = 'gray')
plt.show()
```

# HSV and HLS

```python
img_hsv  =cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
img_hls = cv2.cvtColor(img,cv2.COLOR_BGR2HLS)
fig, (ax1, ax2) = plt.subplots(nrows = 1, ncols = 2, figsize= (20,20))
ax1.imshow(img_hsv)
ax2.imshow(img_hls)
plt.show()
```

# Draw on image

```python
img = cv2.imread('/content/wall.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
display(img)
```

```python
img_copy = img.copy()
cv2.rectangle(img_copy, pt1 =(800,450), pt2 = (980, 550), color = (255,0,0), thickness = 5 )
display(img_copy)
```

```python
cv2.circle(img_copy, center =(500,90), radius = 50, color = (0,0,255), thickness = 5 )
display(img_copy)
```

```python
cv2.putText(img_copy, text= "Have a Nice Day", org = (250, 250), fontFace = cv2.FONT_ITALIC,
            fontScale = 2, color = (0,255,0), thickness = 2, lineType = cv2.LINE_AA)
display(img_copy)
```
